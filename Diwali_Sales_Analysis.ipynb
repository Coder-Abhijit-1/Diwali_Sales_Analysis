```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding = "unicode_escape")
# TO avoid encoding error we have used that unicode_escape word
df
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
# Deleting Null columns
df.drop(["Status","unnamed1"],axis=1, inplace=True)
df
```

```python
pd.isnull(df)
```

```python
# Check NUll values
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
# Few rows are deleted now
df.shape
pd.isnull(df).sum()
```

```python
# For Changing data type use this
df['Amount'] = df['Amount'].astype('int')
# For checking data type use this 
df['Amount'].dtypes
# Now for renaming do this
df.rename(columns = {'Gender':'Sex'}) 
```

```python
ax = sns.countplot(x = 'Gender',data = df)
sns.set(rc = {'figure.figsize':(1,1)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen)
```

```python
ax = sns.countplot(data = df, x = 'Age Group',hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)

sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age)
```

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by= 'Orders',ascending = False)

sns.set(rc = {'figure.figsize':(20,3)})
sns.barplot(x = 'State',y = 'Orders',data = sales_state)
```

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)

sns.set(rc = {'figure.figsize':(20,3)})
sns.barplot(x = 'State',y = 'Amount',data = sales_state)
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status',hue = 'Marital_Status')

sns.set(rc = {'figure.figsize':(1,1)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)

sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status',y = 'Amount',hue = 'Gender',data = sales_status)
```

```python
ax = sns.countplot(data = df, x = 'Occupation')
sns.set(rc = {'figure.figsize':(20,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)

sns.set(rc = {'figure.figsize':(20,3)})
sns.barplot(x = 'Occupation',y = 'Amount',data = sales_occup)
```

```python
sns.set(rc = {'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category',hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_cat = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by= 'Amount',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category',y = 'Amount',data = sales_cat)
```
